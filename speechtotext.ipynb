```python
def text_to_speech(text, lang='en'):
    tts = gTTS(text=text, lang=lang, slow=False)
    tts.save("output.mp3")
    return ipd.Audio("output.mp3")

# Example usage with different languages
text_en = "Hello, welcome to the text-to-speech demo."
text_es = "Hola, bienvenido a la demostración de texto a voz."
text_fr = "Bonjour, bienvenue dans la démonstration de synthèse vocale."

audio_en = text_to_speech(text_en, lang='en')
audio_es = text_to_speech(text_es, lang='es')
audio_fr = text_to_speech(text_fr, lang='fr')

ipd.display(audio_en)
ipd.display(audio_es)
ipd.display(audio_fr)
```

```python
import gradio as gr
from gtts import gTTS
import os
```

```python
# Create the Text-to-Speech Application
def text_to_speech(text, lang='en'):
    tts = gTTS(text=text, lang=lang, slow=False)
    tts.save("output.mp3")
    return "output.mp3"
```

```python
# Define the Gradio interface
iface = gr.Interface(
    fn=text_to_speech,
    inputs=[
        gr.Textbox(label="Enter Text", placeholder="Type here..."),
        gr.Dropdown(choices=["en", "es", "fr", "de", "it", "pt"], label="Select Language")
    ],
    outputs=gr.Audio(label="Generated Audio"),
    title="Text-to-Speech Application",
    description="Enter text and select a language to convert it to speech."
)
```

```python
# Launch the interface
iface.launch(share=True)
```
